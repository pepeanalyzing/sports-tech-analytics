Use Python + pandas + numpy to generate data.
platform_logs.csv, sports_events and user_activity dataset


```python
# Import necessary libraries and generate synthetic data for platform logs, user interactions, and performance metrics. The data will be saved as CSV files for further analysis.
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import os
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# 1. Platform logs
n_logs = 5000
start = datetime(2025, 1, 1)

timestamps = [start + timedelta(minutes=i) for i in range(n_logs)]
endpoints = ['/events', '/login', '/upload', '/analytics']
statuses = ['success', 'error']
error_types = ['timeout', 'api_failure', 'auth_error', 'none']


logs = pd.DataFrame({
    'log_id': range(1, n_logs + 1),
    'timestamp': timestamps,
    'latency_ms': np.random.normal(250, 80, n_logs).clip(50, 1000),
    'endpoint': np.random.choice(endpoints, n_logs),
})

# derive status + error_type
logs['status'] = np.where(np.random.rand(n_logs) < 0.9, 'success', 'error')
logs['error_type'] = np.where(
    logs['status'] == 'error',
    np.random.choice(error_types[:-1], logs.shape[0]),
    'none'
)

logs['date'] = logs['timestamp'].dt.date

logs.to_csv("C:/Users/pepe/sports-tech-analytics/Project 1/data/platform_logs.csv", index=False)

```

🟦 Step 2 — Generate sports_events.csv

This simulates match events: passes, shots, tackles, sprints.

```python


# Ensure folder exists
# os.makedirs("C:/Users/pepe/sports-tech-analytics/Project 1/data", exist_ok=True)

# Generate sports events
n_events = 3000
matches = [f"M{i}" for i in range(1, 21)]
players = [f"P{i}" for i in range(1, 51)]
event_types = ['pass', 'shot', 'tackle', 'sprint']

start = datetime(2025, 1, 1)

events = pd.DataFrame({
    'event_id': range(1, n_events + 1),
    'match_id': np.random.choice(matches, n_events),
    'player_id': np.random.choice(players, n_events),
    'event_type': np.random.choice(event_types, n_events),
    'event_timestamp': [
        start + timedelta(seconds=random.randint(0, 5400))  # 90 min match
        for _ in range(n_events)
    ]
})



events.to_csv("C:/Users/pepe/sports-tech-analytics/Project 1/data/sports_events.csv", index=False)
```

🟦 Step 3 — Generate user_activity.csv

```python
# Generate user activity
n_actions = 4000
users = [f"U{i}" for i in range(1, 101)]
roles = ['coach', 'player', 'analyst']
actions = ['login', 'upload', 'review', 'annotate']

activity = pd.DataFrame({
    'user_id': np.random.choice(users, n_actions),
    'role': np.random.choice(roles, n_actions),
    'action': np.random.choice(actions, n_actions),
    'timestamp': [
        start + timedelta(minutes=random.randint(0, 10000))
        for _ in range(n_actions)
    ]
})

activity.to_csv("C:/Users/pepe/sports-tech-analytics/Project 1/data/user_activity.csv", index=False)
```
